```python
# amestecam randurile
df=df.sample(frac=1)
```

```python
X=df.drop(['Loan_Status'],axis=1).values
y=df['Loan_Status'].values
```

```python
# importam functiile pentru testarea datelor etc
from sklearn.metrics import accuracy_score, plot_confusion_matrix 
from sklearn.model_selection import train_test_split
```

```python
#impartim datele pentru antrenare si testare a datelor
X_train, X_test, y_train, y_test= train_test_split(X,y, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# facem o functie, de a reduce codul scris pana aici
def fit_all(X, y):
    models = [LogisticRegression(max_iter=1000), KNeighborsClassifier(),
              DecisionTreeClassifier(), RandomForestClassifier(),
              GaussianNB()]
    names = ['Logit', 'KNN', 'CART', 'FOREST', 'GAUSS']
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    for i in range(len(models)):
        print(names[i])
        models[i].fit(X_train, y_train)
        y_pred = models[i].predict(X_test)
        y_pred_train = models[i].predict(X_train)
        train_acc = accuracy_score(y_train, y_pred_train)
        test_acc = accuracy_score(y_test, y_pred)
        print(f"train acc - {train_acc}")
        print(f"test acc - {train_acc}")
        plot_confusion_matrix(models[i], X_test, y_test)
```

```python
fit_all(X,y)
```
